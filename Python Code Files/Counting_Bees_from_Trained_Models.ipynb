**Displaying the Bee Counts Neatly**

```python
import os
import random
from PIL import Image, ImageDraw, ImageFont

# location of the exp folder - change as needed
exp = "exp2"

# Set the paths
base_path = "/content/yolov5/inference_results"
labels_path = os.path.join(base_path, exp, "labels")
images_path = os.path.join(base_path, exp, "")

# gather the directory information with labels
all_labels = [label for label in os.listdir(labels_path) if label.endswith(".txt")]

# set the display to be 15 - no need to overload the system and display everything
num_images_to_display = min(15, len(all_labels))
sample_labels = random.sample(all_labels, num_images_to_display)

# Traverse the labels and find the image
for label_name in sample_labels:
    # Extract the image name while removing the the extension
    img_name = os.path.splitext(label_name)[0]

    # set the paths
    label_path = os.path.join(labels_path, label_name)
    img_path = os.path.join(images_path, f"{img_name}.jpg")

    # getting the counts
    num_objects = sum(1 for line in open(label_path))

    # Draw and display the image with the counts
    img = Image.open(img_path)
    draw = ImageDraw.Draw(img)
    font = ImageFont.load_default()  # You can customize the font as needed
    text = f"Bees: {num_objects}"
    draw.text((10, 10), text, fill="red", font=ImageFont.load_default().font_variant(size=36))

    # Display the image with the overlaid text
    display(img)
```
